```python
import pandas as pd
import numpy as np
from pandas_datareader import data
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import yfinance as yf
from datetime import datetime
```

# Base de Dados

```python
acoes = ['ANIM3.SA','COGN3.SA','CSED3.SA','SEER3.SA','YDUQ3.SA', 'BOVA11.SA']
```

```python
acoes_educacao = pd.DataFrame()
for acao in acoes:
  print(acao)
  acoes_educacao[acao] = yf.download(acao, start='2019-01-01')['Close']
```

```python
acoes_educacao = acoes_educacao.rename(columns={'ANIM3.SA': 'ANIMA','COGN3.SA':'KROTON','CSED3.SA':'CRUZEIRO','SEER3.SA':'SER','YDUQ3.SA':'YDUQS','BOVA11.SA':'BOVA'})
```

```python
acoes_educacao.reset_index(inplace=True)
acoes_educacao
```

```python
acoes_educacao.isna().sum()
```

```python
acoes_educacao.describe()
```

```python
acoes_educacao_normalizado = acoes_educacao.copy()
for i in acoes_educacao_normalizado.columns[1:]:
  if i != "CRUZEIRO":
    acoes_educacao_normalizado[i] = acoes_educacao_normalizado[i]/acoes_educacao_normalizado[i][0]
  else:
    acoes_educacao_normalizado[i] = acoes_educacao_normalizado[i]/12.98
```

```python
acoes_educacao_normalizado 
```

```python
figura = px.line(title='Histórico do preço das ações')
for i in acoes_educacao.columns[1:]:
  figura.add_scatter(x=acoes_educacao['Date'], y=acoes_educacao[i], name =i)
figura.show()
```

```python
figura = px.line(title='Histórico do preço das ações - Normalizado')
for i in acoes_educacao_normalizado.columns[1:]:
  figura.add_scatter(x=acoes_educacao_normalizado['Date'], y=acoes_educacao_normalizado[i], name =i)
figura.show()
```
